````python
def quicksort(A, lo, hi):
    def partition(lo, hi):
        pivot = A[hi]
        left = lo
        for right in range(lo, hi):
            if A[right] < pivot:
                A[left], A[right] = A[right], A[left]
                left += 1
        A[left], A[hi] = A[hi], A[left]
        return left
    
    if lo < hi:
        pivot = partition(lo, hi)
        quicksort(A, lo, pivot - 1)
        quicksort(A, pivot + 1, hi)
````

````python
## 입력값 -1 -> 5 -> 3 -> 4 -> 0 연결 리스트에 대한 병합 정렬 과정 도식화
````

````python
class ListNode:
    def __init__(self, val = 0, Next = None):
        self.val = val
        self.next = Next
````

````python
class Solution:    
    # 두 정렬 리스트 병합
    def mergeTwoLists(self, l1: ListNode, l2: ListNode) -> ListNode:
        if l1 and l2:
            if l1.val > l2.val:
                l1, l2 = l2, l1
            l1.next = self.mergeTwoLists(l1.next, l2)

        return l1 or l2

    def sortList(self, head: ListNode) -> ListNode:
        if not (head and head.next):
            return head

        # 런너 기법 활용
        half, slow, fast = None, head, head

        while fast and fast.next:
            half, slow, fast = slow, slow.next, fast.next.next
        half.next = None

        # 분할 재귀 호출
        l1 = self.sortList(head)
        l2 = self.sortList(slow)

        return self.mergeTwoLists(l1, l2)
````

````python
arr = ListNode(-1, ListNode(5, ListNode(0, ListNode(3, ListNode(4)))))

ans = Solution()

a = ans.sortList(arr)
````

##### 풀이 3 내장 함수를 이용하는 실용적인 방법

````python
## 사실 이 문제는 정렬 알고리즘을 직접 구현할 필요가 없다.
## 요즘 대부분의 프로그래밍 언어들은 표준 라이브러리에서 이미 성능 좋은 정렬 알고리즘을 제공하고 있기 때문이다.
## 파이썬에서도 C로 신중하게 구현한 팀소트 내장 함수 .sort()와 sorted()를 제공한다.
## 이는 실행시간을 조금이라도 줄여야 하는 코딩 테스트에서는 가장 현명한 접근 방식이기도 하며, 
## 실무에서도 당연히 이 같은 방식으로 풀이하는 편이 실용적이다.
````

````python
def sortList(head: ListNode) -> ListNode:
    # 연결 리스트 -> 파이썬 리스트
    p = head
    lst = []
    while p:
        lst.append(p.val)
        p = p.next
        
    # 정렬 -> in-place
    lst.sort()
    
    # 파이썬 리스트 -> 연결 리스트
    p = head
    for i in range(len(lst)):
        p.val = lst[i]
        p = p.next
    return head
````

````python
from typing import *

def merge(intervals: List[List[int]]) -> List[List[int]]:
    merged = []
    for i in sorted(intervals, key = lambda x: x[0]):
        if merged and i[0] <= merged[-1][1]:
            merged[-1][1] = max(merged[-1][1], i[1])
        else:
            merged += i,
            
    return merged
````

````python
## 찾은 cur 위치 다음에 head가 들어가고, head.next에는 cur.next를 연결해 계속 이어지게 한다.
## 그리고 다음번 head는 head.next로 차례를 이어받는다.
## 이후에는 cur = parent를 통해 다시 처음으로 되돌아가며, 차례대로 다시 비교하게 된다.
````

````python
def insertionSortList(head: ListNode) -> ListNode:
    cur = parent = ListNode(None)
    
    while head:
        while cur.next and cur.next.val < head.val:
            cur = cur.next
            
        cur.next, head.next, head = head, cur.next, head.next
        
        cur = parent
        
    return cur.next
````

````python
## 딱 한 줄의 조건문을 추가해봤다.
## 수정 사항을 반영한 전체 코드는 다음과 같다.
````

````python
def insertionSortList(head: ListNode) -> ListNode:
    cur = parent = ListNode(None)
    
    while head:
        while cur.next and cur.next.val < head.val:
            cur = cur.next
            
        cur.next, head.next, head = head, cur.next, head.next
        
        if head and cur.val > head.val:
            cur = parent
        
    return cur.next
````

````python
## 예제 1
    ## input: [10, 2]
    ## output: "210"
    
## 예제 2
    ## input: [3, 30, 34, 5, 9]
    ## output: "9534330"
````

##### 풀이 1 삽입 정렬

````python
## 어떻게 풀이해야 할까?
## 각 요소 단위로 크기 순으로 정렬하면 된다.
## 단, 여기서는 다소 특이한 정렬 기법을 적용해야 하는데 맨 앞에서부터 자릿수 단위로 비교해서 크기 순으로 정렬한다.

## 즉 9는 30보다 맨 앞자리 수가 더 크므로 9가 더 앞에 와야 한다.
## 좀 더 쉽게 비교하려면 930이 큰지 309가 큰지 비교하는 문제로 풀 수 있을 것 같다.
````

```python
def to_swap(n1: int, n2: int) -> bool:
    return str(n1) + str(n2) < str(n2) + str(n1)
```

````python
## 이 함수의 결과가 True라면 위치 변경이 이뤄져야 한다.

## 앞서 삽입 정렬을 구현해봤는데, 앞의 풀이는 연결 리스트에 대한 풀이이며, 
## 이 문제는 배열을 기반으로 한 삽입 정렬을 구현해 문제를 풀어본다.
## 대부분의 알고리즘 책에서도 배열 풀이를 기본으로 하고 있으므로 이 문제 또한 새롭게 배열로 구현하는 형태로 풀이해보자.
````

````python
class Solution:
    # 문제에 적합한 비교 함수
    @staticmethod
    def to_swap(n1: int, n2: int) -> bool:
        return str(n1) + str(n2) < str(n2) + str(n1)
    
    # 삽입 정렬 구현
    def largestNumber(self, nums: List[int]) -> str:
        i = 1
        while i < len(nums):
            j = i
            while j > 0 and self.to_swap(nums[j - 1], nums[j]):
                nums[j], nums[j - 1] = nums[j - 1], nums[j]
                j -= 1
            i += 1
        
        return str(int(''.join(map(str, nums))))
````

````python
## 마지막에 리턴하는 부분이 다소 번잡하다.
## 원래 ''.join(map(str, nums)) 정도면 끝나겠지만, 여기서는 입력값이 ["0", "0"] 인 경우도 있기 때문에
## 그냥 문자로 처리해버리면 리턴값이 00이 된다.
## 따라서 join 결과를 int로 바꿔서 00이 0이 되도록 만들어 준 후, 다시 str로 변경해서 처리한다.
````

````python
## 배열은 이처럼 깔끔하게 구현할 수 있기 때문에 연결 리스트 때처럼 포인터를 맨 앞으로 돌릴지 비교하는 등의 
## 최적화는 수행할 필요가 없다. 
## 또 정렬된 리스트 변수를 별도로 선언했던 것과 달리 배열로 구현하게 되면, 
## 이처럼 제자리 정렬(In-Place Sort)이 가능하여 공간 복잡도도 줄일 수 있다.
````

````python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,15))
columns = 2
images = ["리스트17-5.jpg", "그림17-12.jpg"]
for i, path in enumerate(images):
    image = plt.imread(path)
    plt.subplot(len(images) / columns + 1, columns, i + 1)
    plt.imshow(image)
````

````python
## 이 수도코드는 i, k를 양쪽 포인터로 두고 j가 이동하면서 mid 값을 기준으로 스왑하는 형태로 구현되어 있다.
## 좀 더 직관적으로 풀이하기 위해 변수명을 국기 색깔에 대입해 
## i를 red로, k를 blue로, j를 white로 하고, 중앙값을 지칭하는 mid는 원래의 값인 1로 변경해보자.
## 예제의 입력값에서 보듯 각각의 실제 값은 [0, 1, 2]로 정한다.

## 시작 시점에는 red, white 모두 0에서 시작하고 blue는 배열의 길이가 된다.
## 배열은 0부터 시작하므로 처음에 blue는 배열 인덱스의 바깥에 있다.

## 비교가 시작되면서 1을 기준으로 작은 값은 왼쪽으로, 큰 값은 오른쪽으로 스왑된다.
## 이때 red와 blue는 각각 오른쪽과 왼쪽으로 이동하면서 간격이 점점 더 좁아지는 형태가 된다.
## 비교가 완료될 때 red는 1보다 작은 인덱스의 +1지점, blue는 1보다 큰 인덱스의 처음을 가리키게 되며,
## white와 blue가 겹치면서 비교가 완료된다.
````

````python
def sortColors(nums: List[int]) -> None:
    red, white, blue = 0, 0, len(nums)
    
    while white < blue:
        if nums[white] < 1:
            nums[red], nums[white] = nums[white], nums[red]
            white += 1
            red += 1
        elif nums[white] > 1:
            blue -= 1
            nums[white], nums[blue] = nums[blue], nums[white]
        else:
            white += 1
````

````python
nums = [2, 0, 2, 1, 1, 0]

sortColors(nums)
````

# 64. 원점에 K번째로 가까운 점

### LeetCode 973. K Closest Points to Origin

- 평면상에 points 목록이 있을 때, 원점 (0, 0)에서 가장 가까운 점 K개 목록을 순서대로 출력하라.
- 평면상의 두 점의 거리는 유클리드 거리로 한다.

````python
## 예제 1
    ## input: points = [[1, 3], [-2, 2]], K = 1
    ## output: [[-2, 2]]
    
    ## 설명: (1, 3)과의 거리는 √10 이고, (-2, 2)와의 거리는 √8 이다. 
    ##       두 번째가 더 가까우며, 가장 가까운 거리 1개는 [[-2, 2]]이다.
    
## 예제 2
    ## input: points = [[3, 3], [5, -1], [-2, 4]]
    ## output: [[3, 3], [-2, 4]]
````

````python
## 유클리드 거리(Euclidean Distance)는 유클리드 공간에서 두 점 사이의 거리를 계산하는 가장 '일반적인 방법' 중 하나다.

## 예를 들어 원점 (0, 0)을 x 좌표라고 할 때, (1, 3)에 있는 y 좌표와의 유클리드 거리는 √10이 된다.
## 이 값을 크기가 작은 순으로 K번 추출하면 되는 어렵지 않게 풀 수 있는 문제다.

## K번 추출이라는 단어에서 바로 우선순위 큐를 떠올릴 수 있다.
## 즉 유클리드 거리를 계산하고 이 값을 우선순위 큐로 K번 출력하면 쉽게 문제를 풀이할 수 있다.
````

````python
import heapq

def KClosest(points: List[List[int]], K: int) -> List[List[int]]:
    heap = []
    for (x, y) in points:
#         dist = math.sqrt((0 - x) ** 2 + (0 - y) ** 2)
        dist = x ** 2 + y ** 2
        heapq.heappush(heap, (dist, x, y))
        
    result = []
    for _ in range(K):
        (dist, x, y) = heapq.heappop(heap)
        result.append((x, y))
    return result
````

````python
points = [[3, 3], [5, -1], [-2, 4]]

KClosest(points, 2)
````

````python
## 계산한 거리를 다른 데서 활용할 게 아니라면 굳이 math.sqrt를 쓸 필요없다.
## 아차피 순서만 동일하면 되기 때문이다.
## 이 경우 불필요한 계산을 줄일 수 있기 때문에 수행 속도를 조금 더 높일 수 있다.
````
